```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image_path = "data/STSH/frame_34.jpg"
img = cv2.imread(image_path)
```

```python
img = cv2.imread(image_path)

# Lấy kích thước ảnh
h, w = img.shape[:2]

# Xác định vùng logo (trên cùng bên phải)
logo_height = 200
logo_width = 300
x_start = w - logo_width
y_start = h - logo_height
```

```python
# thử để xác định tọa độ vùng chưa logo 
# Tạo mask cho vùng logo
mask = np.zeros(img.shape[:2], dtype="uint8")
mask[y_start:y_start+logo_height - 30, x_start:w - 60] = 255  # vùng logo = 255

# Inpainting để lấp vùng bị xóa bằng nội dung xung quanh
img_inpaint = cv2.inpaint(img, mask, 3, cv2.INPAINT_TELEA)

# cv2.imwrite("image_no_logo_inpaint.jpg", img_inpaint)

plt.imshow(cv2.cvtColor(img_inpaint, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
import cv2
import numpy as np
import os
from tqdm import tqdm
```

```python
def remove_logo_right(input_folder, output_folder, logo_width, logo_height):
    """
    input_folder: thư mục chứa ảnh .jpg
    output_folder: nơi lưu ảnh đã xử lý
    logo_width: chiều rộng logo (pixel)
    logo_height: chiều cao logo (pixel)
    """
    os.makedirs(output_folder, exist_ok=True)

    for filename in tqdm(os.listdir(input_folder)):
        if filename.lower().endswith(".jpg"):
            img_path = os.path.join(input_folder, filename)
            img = cv2.imread(img_path)
            if img is None:
                continue

            h, w = img.shape[:2]
            mask = np.zeros((h, w), dtype="uint8")

            x_start = 40
            y_start = 0
            mask[y_start:y_start+logo_height, x_start: x_start + logo_width] = 255

            # Inpainting
            img_inpaint = cv2.inpaint(img, mask, 3, cv2.INPAINT_TELEA)

            out_path = os.path.join(output_folder, filename)
            cv2.imwrite(out_path, img_inpaint)

    print(f"✅ Xử lý xong, ảnh lưu ở: {output_folder}")
```

```python
# Chạy cho thư mục
remove_logo_right(
    input_folder="data/STSH",
    output_folder="data/STSH_no_logo",
    logo_width=180,
    logo_height=100
)
```
